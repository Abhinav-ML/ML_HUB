Using facetgrid we can create like multiple small little graphs based on categories.
And this is very interesting and this is very useful and rather powerful concept in order to generate
multiple graph based on categories.let's understand how we can do that.
So first of all, we will be passing the data then we need to pass the col values and row values, so row values 
and column values will be based on the future.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='ticks')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
grid = sns.FacetGrid(data = tips,row ='smoker',col='day',col_order=['Sun','Sat','Fri','Thur'])\
        .map(sns.histplot,'total_bill')
```

# facetgrid as Scatter plot

```python
grid = sns.FacetGrid(data = tips,col='day',col_wrap =2)\
        .map(sns.scatterplot,'total_bill','tip')
# So if you have a long line of code along the line of statement, you can break that like this, you can simply type 
# in backward slash and use the next line to continue with your statement.
plt.show()
```
