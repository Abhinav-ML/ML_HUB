```python
q = data['Pregnancies'].quantile(0.98)
# we are removing the top 2% data from the Pregnancies column
data_cleaned = data[data['Pregnancies']<q]
print(data_cleaned.shape)
q = data_cleaned['BMI'].quantile(0.99)
# we are removing the top 1% data from the BMI column
data_cleaned  = data_cleaned[data_cleaned['BMI']<q]
print(data_cleaned.shape)
q = data_cleaned['SkinThickness'].quantile(0.99)
# we are removing the top 1% data from the SkinThickness column
data_cleaned  = data_cleaned[data_cleaned['SkinThickness']<q]
print(data_cleaned.shape)
q = data_cleaned['Insulin'].quantile(0.95)
# we are removing the top 5% data from the Insulin column
data_cleaned  = data_cleaned[data_cleaned['Insulin']<q]
print(data_cleaned.shape)
q = data_cleaned['DiabetesPedigreeFunction'].quantile(0.99)
# we are removing the top 1% data from the DiabetesPedigreeFunction column
data_cleaned  = data_cleaned[data_cleaned['DiabetesPedigreeFunction']<q]
print(data_cleaned.shape)
q = data_cleaned['Age'].quantile(0.99)
# we are removing the top 1% data from the Age column
data_cleaned  = data_cleaned[data_cleaned['Age']<q]
print(data_cleaned.shape)
```

```python
data_cleaned.head(2)
```

# Seperating the features

```python
X = data_cleaned.drop(columns = ['Outcome'])
X.head()
X.shape
```

```python
y = data_cleaned['Outcome']
y.head()
y.shape
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
X_scaled = scalar.fit_transform(X)
```

```python
X_scaled
```

# Train - Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size = 0.25,random_state = 10) 
# Order of this X_train,X_test,y_train,y_test  matter because here we simply assigning  Many Values to Multiple Variables
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Model Building

```python
# Import LogisticRegression class from sklearn
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
y_predict.shape
y_test.shape
```

## Accuracy Score

```python
# Now checking Accuracy Score
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_predict)
accuracy
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
conf_mat = confusion_matrix(y_test,y_predict) # in parameter 1st will y_true and in 2nd y_pred
conf_mat
```

```python
True_Positive = conf_mat[0][0]
True_Positive 
```

```python
False_Positive = conf_mat[0][1]
False_Positive
```

```python
False_Negative = conf_mat[1][0]
False_Negative
```

```python
True_Negative = conf_mat[1][1]
True_Negative
```

```python
# Accuracy using formula
value = (True_Positive + True_Negative)/(True_Positive+False_Positive+False_Negative+True_Negative)
```

```python
value
```

```python
# Precision

Precision = True_Positive/(True_Positive + False_Positive)
Precision
```

```python
# Recall
Recall = True_Positive/(True_Positive+False_Negative)
Recall
```

```python
# F1 Score
F1 = 2*(Precision * Recall )/(Precision + Recall)
F1
```
