```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
names = ['sepal_length','sepal_width','petal_length','petal_width','class']
iris = pd.read_csv('iris.data',names=names)
```

```python
iris.head()
```

```python
sns.FacetGrid(data=iris,hue ='class').map(sns.histplot,'sepal_length').add_legend()
plt.show()
```

# Boxplot

```python
sns.boxplot(data=iris,x='petal_length',y='class')
plt.show()
```

# Violin Plot

```python
sns.violinplot(data=iris,x='petal_length',y='class')
fig = plt.gcf()
fig.set_size_inches(12,15)
plt.show()
```

# heatmap and correlationship

```python
iris.corr()
```

```python
sns.heatmap(iris.corr())
plt.show()
```
