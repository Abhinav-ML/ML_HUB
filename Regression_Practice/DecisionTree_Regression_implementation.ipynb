```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('canada_per_capita_income.csv')
```

```python
data.head()
```

```python
# checking the relationship between data is linear or not
# sns.scatterplot(x = 'per capita income (US$)', y = 'year',data = data)
# plt.show()
```

```python
# checking whether data is having null value or not
# pd.isnull(data)
pd.isnull(data).sum()
```

# Separating feature and target value

```python
X = data.iloc[:,:-1]
# type(X)
```

```python
Y = data.iloc[:,-1]
```

# Train-Test Split

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state = 1)
```

# Building Model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_depth = 10)
```

```python
model.fit(X_train,y_train)
```

# Making the prediction

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

# Comparison

```python
comparison = pd.DataFrame()
```

```python
comparison['Actual'] = y_test
```

```python
comparison['Predicted'] = y_pred
```

```python
comparison
```

# Evaluation

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```
