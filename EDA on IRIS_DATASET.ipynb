# Iris DataSet


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.set()
# When we plan to use Seaborn in order to get better view of matplotlib 
#we need to set seaborn(in other words need to call sns.set())
```

```python
header = ['sepal_length','sepal_width','petal_length','petal width','class'] 
#in order to get header
```

```python
iris = pd.read_csv('iris.data',names = header)
iris.head()
```

```python
# in oreder to check the count of class
iris['class'].value_counts()

# iris['class'].nunique() #checking total no of unique class 
```

# 2D Scatter Plot

```python
# we are now making 2D plot on basis of sepal_length and sepal_width
ax = sns.scatterplot(x='sepal_length',y ='sepal_width',data = iris,hue='class')
plt.show()
```

# **** Pair Plot *****

```python
sns.pairplot(data= iris,hue='class')
plt.show()
```
